```python
plt.style.use('seaborn')
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')

fig.subplots_adjust(bottom=0, top=1, left=0, right=1)

ax.set_xlabel('X Axis', fontsize=20, labelpad=20)
ax.set_ylabel('Y Axis', fontsize=20, labelpad=20)
ax.set_zlabel('Z Axis', fontsize=20, labelpad=20)

PI = np.pi
n_point = 300

t = np.linspace(0, 10*PI, n_point)
x = t*np.cos(t)
y = t*np.sin(t)
z = 0.2*t

ax.plot(x, y, z, color='r', alpha=0.7, linestyle='-', linewidth=3)

zlim = ax.get_zlim()
ax.plot(x, y, zlim[0]*np.ones(n_point), color='gray')

ylim = ax.get_ylim()
ax.plot(x, ylim[0]*np.ones(n_point), z, color='gray')

xlim = ax.get_xlim()
ax.plot(xlim[0]*np.ones(n_point), y, z, color='gray')
```

```python
plt.style.use('seaborn')
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')

fig.subplots_adjust(bottom=0, top=1, left=0, right=1)

ax.set_xlabel('X Axis', fontsize=20, labelpad=20)
ax.set_ylabel('Y Axis', fontsize=20, labelpad=20)
ax.set_zlabel('Z Axis', fontsize=20, labelpad=20)

PI = np.pi
n_point = 300

t = np.linspace(0, 10*PI, n_point)
x = t*np.cos(t)
y = t*np.sin(t)
z = 0.2*t

import matplotlib.cm as cm

s_arr = np.linspace(10, 100, n_point)
cmap = cm.get_cmap('rainbow', lut=n_point)
c_arr = [cmap(c_idx) for c_idx in range(n_point)]
ax.scatter(x, y, z, s=s_arr, c=c_arr)
```

```python
plt.style.use('seaborn')
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')

fig.subplots_adjust(bottom=0, top=1, left=0, right=1)

ax.set_xlabel('X Axis', fontsize=20, labelpad=20)
ax.set_ylabel('Y Axis', fontsize=20, labelpad=20)
ax.set_zlabel('Z Axis', fontsize=20, labelpad=20)

PI = np.pi
n_point = 300

t = np.linspace(0, 10*PI, n_point)
x = t*np.cos(t)
y = t*np.sin(t)
z = 0.2*t

noise_coeff = 0.1
x_noise = x + noise_coeff*np.random.normal(0, 1, n_point)
y_noise = y + noise_coeff*np.random.normal(0, 1, n_point)
z_noise = z + noise_coeff*np.random.normal(0, 1, n_point)

ax.scatter(x_noise, y_noise, z_noise, s=100)
ax.plot(x, y, z, color='r', linewidth=1)

def plot_text(ax, x, y, z, annot, color):
    ax.scatter(x, y, z, s=500, color=color)
    
    ax.text(x, y, z, annot, color=color, fontsize=30)
    
plot_text(ax, x_noise[0], y_noise[0], z_noise[0], 'Start Point', 'b')
plot_text(ax, x_noise[0], y_noise[150], z_noise[150], 'Middle Point', 'g')
plot_text(ax, x_noise[0], y_noise[-1], z_noise[-1], 'End Point', 'r')
```

```python
plt.style.use('seaborn')
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)

X, Y = np.meshgrid(x, y)

coordinates = np.empty(X.shape, dtype=object)

for r_idx in range(X.shape[0]):
    for c_idx in range(X.shape[1]):
        x = X[r_idx, c_idx]
        y = Y[r_idx, c_idx]
        
        coordinates[r_idx, c_idx] = (x, y)
        
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(projection='3d')

ax.tick_params(labelsize=20)
fig.subplots_adjust(bottom=0, top=1, left=0, right=1)

surface = ax.plot_surface(X, Y, Z, cmap='hot')
cbar = fig.colorbar(surface, orientation='horizontal', fraction=0.1, pad=0, shrink=0.9)
```
