```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
np.random.seed(0)

data = np.array([1, 2, 3, 4, 5])
data_idx = np.arange(len(data))

fig, axes = plt.subplots(1, 2, figsize=(14, 7))
axes[0].tick_params(labelsize=15)
axes[1].tick_params(labelsize=15)

axes[0].bar(data_idx, data)
axes[1].barh(data_idx, data)
```

```python
np.random.seed(0)

n_data = 10
data = np.random.uniform(10, 100, (n_data))
data_idx = np.arange(n_data)
data_label = ['Class ' + str(i+1) for i in range(n_data)]

fig, ax = plt.subplots(figsize=(10, 7))
ax.invert_yaxis()
ax.barh(data_idx, data)
ax.set_yticks(data_idx)
ax.set_yticklabels(data_label, rotation=30, va='top')
# va=top 좌표가 text 위에 놓임
```

```python
# Exercise 4-01

### data setting
xticklabels = np.array(['White', 'Black', 'Gray', 'Red', 'Green', 'Blue', 'Yello', 'Brown', 'Other'])

xticks = np.arange(len(xticklabels))

data = np.array([47.19, 15.82, 14.32, 7.30, 2.64, 5.64, 0.84, 5.38, 0.73])
background = 55*np.ones_like(data)

fig, ax = plt.subplots(figsize=(20, 10))

### background plotting
ax.bar(xticks, background, color='k', alpha=0.05, hatch='/', edgecolor='k')

### bar plotting
rects = ax.bar(xticks, data, color='cornflowerblue')

### tick & ticklabel customizing
ax.set_xticks(xticks)
ax.set_xticklabels(xticklabels, fontsize=50, rotation=55)

ax.tick_params(axis='y', labelsize=0, left=False)
ax.tick_params(axis='x', bottom=False)

### spine customizing
for spine in ax.spines.values():
    spine.set_visible(False)
    
### annotation
for rect in rects:
    height = rect.get_height()
    width = rect.get_width()
    x = rect.get_x()
    
    ax.text(x + width/2, 55, 
           str(height) + '%',
           fontsize=40, rotation=-55, ha='right', va='bottom')
```
