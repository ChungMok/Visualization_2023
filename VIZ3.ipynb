```python
import numpy as np
import csv

def bop_data_reader():
    with open('C:/Users/Admin/OneDrive/ /Sungkyun University//DScover (2023)/ /9th_siltae_public_new_211102.csv') as csvfile :
        csvreader = csv.reader(csvfile)
        next(csvreader)

        dataset = list()
        month_list = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
        month_dict = {month_list[m_idx]:m_idx+1 for m_idx in range(len(month_list))}

        for data in csvreader:
            price = data[-1]
            date = data[0]
            try:
                D, M, Y = date.split(sep='-')
                if int(Y) >= 87:
                    M = month_dict[M]
                    dataset.append([Y, M, D, price])
            except:
                pass

    dataset = np.array(dataset).astype(np.float)
    return dataset

def get_year_data(dataset, t_year):
    t_idx = np.where(dataset[:,0] == t_year)
    t_data = dataset[t_idx]
    return t_data
```

```python
from bop_utils import *

dataset = bop_data_reader()
t_year = 90
t_year_data = get_year_data(dataset, t_year)

fig, ax = plt.subplots(figsize=(15, 10))
ax.plot(t_year_data[:, -1])

ax.set_title('BOP data({})'.format(t_year), fontsize=30)
ax.set_ylabel("daily prices in USD", fontsize=25)
ax.tick_params(labelsize=20)
ax.grid(axis='y')
```
