```python
PI = np.pi
t = np.linspace(-4*PI, 4*PI, 200)
sin = np.sin(t)

fig, ax = plt.subplots(figsize=(10, 7))
ax.plot(t, sin, color='black')
ax.axhline(0, color='black')
ax.fill_between(t, sin, facecolor='white', hatch='/', edgecolor='b', alpha=0.3)
```

```python
# Exercise 02-05

### data setting
name_list = ['PL', 'NL', 'NL-<SeINL', 'NL->SeINL->SeIPL']

x_data = np.linspace(0.5, 1, 30)
y_data = np.empty(shape = (0, len(x_data)))
gamma_list = [6, 7, 8, 12]
for i in range(len(gamma_list)):
    y = (-0.5)*np.power(x_data, gamma_list[i]) + 1
    y = y.reshape(1, len(x_data))
    
    y_data = np.vstack((y_data, y))
    
### plotting
fig, ax = plt.subplots(figsize=(12, 12))
for i in range(len(name_list)):
    ax.plot(x_data, y_data[i],
           label=name_list[i],
           marker='o',
           markersize=10)

### spine customizing
for spine_loc, spine in ax.spines.items():
    spine.set_linewidth(3)
    spine.set_alpha(0.5)
    
### axis limit customizing
ax.set_xlim([0.5, 1.0])
ax.set_ylim([0.5, 1.0])

### legend customising 
ax.legend(loc='lower left', fontsize=17)

### tick & ticklabels customizing
tick_dict = {'size':15}
ticks = [round(0.1*i, 1) for i in range(5, 11)]

ax.set_xticks(ticks)
ax.set_xticklabels(ticks, fontdict=tick_dict)
ax.set_yticklabels(ticks, fontdict=tick_dict)

### lable cutomizing
label_dict = {'size':20}
ax.set_xlabel('recall', fontdict=label_dict)
ax.set_ylabel('precision', fontdict=label_dict)

### grid customizing
ax.grid(linewidth=3, alpha=0.5)
```
